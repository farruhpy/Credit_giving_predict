```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("/content/train_u6lujuX_CVtuZ9i (1).csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.dropna()
```

```python
df.info()
```

```python
df.replace({'Loan_Status':{'N':0, 'Y':1}}, inplace=True)
```

```python
df.head()
```

```python
df.replace(to_replace='3+', value=4, inplace=True)
```

```python
df.Dependents.value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Education', hue='Loan_Status', data=df)
```

```python
df.replace({"Married":{'No':0, "Yes":1}, 'Gender':{"Male":0, "Female":1}, "Self_Employed":{"No":0, "Yes":1},
            "Education":{"Not Graduate": 0, "Graduate":1},
            "Property_Area":{"Urban":2, "Semiurban":1, "Rural":0}}, inplace=True)
```

```python
df
```

```python
X = df.drop(columns=['Loan_Status', "Loan_ID"])
Y = df['Loan_Status']
```

```python
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X , Y,  test_size=0.1, random_state=3, stratify=Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
### Training the model
from sklearn import svm
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
### Accuracy test
```

```python
from sklearn.metrics import accuracy_score
```

```python
X_train_prediction = classifier.predict(X_train)
```

```python
training_data_acc=accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy : ", training_data_acc)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_acc=accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy : ", test_data_acc)
```

As you can see our ML model's accuracy isn't that bad: ~80%
